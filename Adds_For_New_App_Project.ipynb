# Limpeza e separação de dados

### Verificar se há linhas duplicadas
É necessário verificar se existem duplicatas em nosso conjuntos de dados, afinal, não é interessante contrar o mesmo dado duas ou mais vezes. 

Para fazer essa verificação criarei uma função para analisar o nome dos aplicativos dentro de uma lista de listas

```python
### Função para verificar nomes duplicados dentro dos conjuntos de dados ###
def verify_duplicates(dataset):
    duplicate_apps = []
    unique_apps = []

    for app in dataset:
        name = app[0]
        if name in unique_apps:
            duplicate_apps.append(name)
        else:
            unique_apps.append(name)
    
    print('Número de aplicativos duplicados:', len(duplicate_apps))
    print('\n')
    print('Examples de aplicativos duplicados:', duplicate_apps[:15])
    
verify_duplicates(ios) # 0 nomes duplicados
print('\n')
verify_duplicates(android) # 1.181 nomes duplicados
```

```python
### Análisando duplicatas ###
for app in android:
    name = app[0]
    if name == 'Quick PDF Scanner + OCR FREE':
        print(app)
```

### Separando duplicatas
Algumas duplicatas são exatamente iguais porém vê-se alguma diferença na quarta coluna que corresponde ao numero de avaliações.

Não deve-se excluir linhas aleatóriamente pois existe a chance do numero de avaliações mais alto pode ser o mais atual. Quero manter a duplicata com o maior numero de avaliações e descartar as outras.

```python
### Pegando apenas as duplicatas com mais avaliações ###
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
print('Comprimento esperado:', len(android) - 1181)
print('Comprimento atual:', len(reviews_max)) # 9.659 linhas de comprimento
```

### Separação: Apenas valores únicos

```python
### Dicionário reviews_max para criar uma lista sem duplicatas ###
android_clean = []
already_added = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

```python
### Verificando comprimento da nova lista ###
explore_data(android_clean, 0, 1, True) # 9.659 linhas de comprimento.
```

### Apps de língua: English
Existem alguns aplicativos que não são destinados ao público falante de inglês, eis alguns exemplos:

```python
print(ios[813][1])
print(ios[6731][1])

print(android_clean[4412][0])
print(android_clean[7940][0])
```

Não existe interesse em mantes esse tipo de aplicativos não destinados aos falantes de Inglês então será necessário separá-los dos dados que desejamos análisar. 

Sabe-se que cada caractere de uma string corresponde a um índice(número) então para facilitar a separação deve-se separar os caracteres que são comuns no idioma inglês das que não são. 

Também existem nomes de aplicativos com alguns caracteres diferentes, para não perder dados colocaremos um limite de 3 caracteres diferentes do normal por nome.

**Segundo a ASCII os índices dos caracteres que desejamos estão entre os índices 0 e 127.**

```python
### Função para verificar índices dos caracteres com limite de 3 Falses ###
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

print(is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('Instachat 😜'))
```

```python
### Pegamos listas sem arquivos duplicados e verificaremos se is_english ###
android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)
        
for app in ios:
    name = app[1]
    if is_english(name):
        ios_english.append(app)
        
explore_data(android_english, 0, 1, True)
print('\n')
explore_data(ios_english, 0, 1, True)
```

# Análise: Visualizando a Superfície
nosso objetivo é determinar os tipos de aplicativos que provavelmente atrairão mais usuários, porque o número de pessoas que usam nossos aplicativos afeta nossa receita.

Como nosso objetivo final é adicionar o aplicativo ao Google Play e à App Store, precisamos encontrar perfis de aplicativos que sejam bem-sucedidos em ambos os mercados.

```python
### Função para encontrar a frequencia de uma coluna numa lista de listas em porcentagem ###
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        percentage = round(percentage, 2)
        table_percentages[key] = percentage 
    
    return table_percentages

### Função para ordenar a função acima ###
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(ios_final, -5) # Coluna 'prime_genre' porcentagem
```

Podemos observar que, entre os aplicativos gratuitos em inglês, **mais da metade (58,16%) são jogos**. Aplicativos de entretenimento representam cerca de 8%, seguidos por aplicativos de fotos e vídeos, que representam cerca de 5%. Apenas 3,66% dos aplicativos são voltados para educação, seguidos por aplicativos de redes sociais, que representam 3,29% dos aplicativos em nosso conjunto de dados.

A impressão geral é que **na App Store (pelo menos a parte que contém aplicativos gratuitos em inglês) é dominada por aplicativos voltados para diversão (jogos, entretenimento, fotos e vídeos, redes sociais, esportes, música, etc.)**, enquanto aplicativos com propósitos práticos (educação, compras, serviços públicos, produtividade, estilo de vida, etc.) são mais raros. No entanto, o fato de os aplicativos de diversão serem os mais numerosos não significa que eles também tenham o maior número de usuários — a demanda pode não ser igual à oferta.

```python
display_table(android_final, 1) # Coluna 'Category' porcentagem
```

O cenário parece significativamente diferente **no Google Play**: não há muitos aplicativos projetados para diversão, e parece que **um bom número deles é projetado para fins práticos (família, ferramentas, negócios, estilo de vida, produtividade, etc.)**. No entanto, se investigarmos isso mais a fundo, veremos que a **categoria família (que representa quase 19% dos aplicativos) inclui principalmente jogos para crianças**.

## Apps da App Store com mais usuários
O conjunto de dados da App Store não tem uma coluna especifica para dizer a quantidade de instalações por aplicativo. Levando em consideração que para se avaliar um aplicativo deve-se tê-lo usado/instalado antes, irei usar o total de avaliações para nos dar uma base sobre a média de instalações por gênero.

```python
### Encontrando a média de instalações por genero na App Store ###
genres_ios = freq_table(ios_final, -5)
for genre in genres_ios:
    total_ratings = 0
    len_genre = 0
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[5])
            total_ratings += n_ratings
            len_genre += 1
    avg_n_ratings = total_ratings / len_genre
    avg_n_ratings = round(avg_n_ratings, 2)
    print(genre, ':', avg_n_ratings)
```

## Apps da Google Play com mais usuários

```python
### Encontrando a média de instalações por genero no Google Play ###
category_android = freq_table(android_final, 1)
for category in category_android:
    total_installs = 0
    len_category = 0
    for app in android_final:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            n_installs = float(n_installs)
            total_installs += n_installs
            len_category += 1
    avg_n_ratings = total_installs / len_genre
    avg_n_ratings = round(avg_n_ratings, 2)
    print(category, ':', avg_n_ratings)
```

A Maiores médias entre o numero de aplicativos e quantidade de downloads na App Store são principalmente: **'Navigation', 'Reference' e 'Social Networking'**.

A Maiores médias entre o numero de aplicativos e quantidade de downloads no Google Play são principalmente: **'Communication', 'Video Players', 'Social','Productivity', 'Photography', 'Game', 'Travel and Local', 'Entertainment', e 'Books and Reference' **.
